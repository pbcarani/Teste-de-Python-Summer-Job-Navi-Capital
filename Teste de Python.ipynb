### 1. Fazer um algoritmo para me dizer quantos números Pares, Múltiplos de 49 e Múltiplos de 37 (Simultaneamente, ou seja, o número deve satisfazer essas 3 condições) existem no intervalo de 1 a 5.000.000 (5 milhões)

```python
#criando uma variavel de contagem
contagem=0
#criando o intervalo determinado pelo enunciado
intervalo = range (1, 5000000)
#criando um loop
for numero in intervalo:
    #verificando se o numero é par
    if numero % 2 == 0:
        #verificando se o numero é multiplo de 49
        if numero % 49 == 0:
            #verificando se o numero é multiplo de 37
            if numero % 37 == 0:
                #somando o numero na contagem
                contagem+=1

print('Existem {} números Pares, múltiplos de 49 e múltiplos de 37'.format(contagem))
```

### 2. Fazer um algoritmo para criar um vetor x com 10 posiçōes começando pelo zero. A construção desse vetor dará pela seguinte maneira 

### Sendo "i" a posição do elemento dentro desse vetor x

### Se a posição for par, então esse elemento na posição i do vetor x deverá ser alimentado da seguinte maneira:

                        X[i]=3ˆi + 7*(i!)

### Lembrando que "i" indica o fatoria da posição i.

### Caso o i seja um valor ímpar, então esse elemento na posição i do vetor x devera ser alimentado da seguinte maneira

                        X[i] = 2^i + 4*ln(i)
                        
### Lembrando que "ln(i)" indica o logaritmo neperiano do número i

### Alimentando esse vetor, deseja saber qual a posição do maior elemento desse vetor e qual a média dos elementos contidos nesse vetor (Arredonde o valor para 2 casas decimais).

    

```python
#importando a biblioteca numpy como np
import numpy as np
#criando variavel de qtd de posições
posicoes=10
#criando o vetor
vetor=[0]*posicoes
#contador para fatorial
n=0
#criando o loop para a construcao do vetor
for i in range(0,posicoes):
    #condicao se a posicao for par
    if i%2 ==0:
        vetor[i] = 3**i+7*(i*n-1)*n
    #condicao se a posicao for impar
    else:
        vetor[i] = 2**i+4*np.log(i)
    n+=1
#valor máximo
maximo=max(vetor)
#média
media=sum(vetor) / len(vetor)
print ("Posição do maior elemento:",vetor.index(maximo))
print ("Média dos elementos contidos nesse valor: {:.2f}".format(media))
```

### 3. Fazer um algoritmo que leia a nota de 5 alunos e uma nota para cada aluno. Guarde essas informações em um dicionário e depois apresente o aluno com a maior nota e sua respectiva nota.

```python
#criando um dicionario
dicionario={}
#lista com os nomes do Alunos
alunos=["Aluno 1","Aluno 2","Aluno 3","Aluno 4","Aluno 5"]
#listas com as notas 
notas=[4,7,9,5,3]
#contador
c=0
#lendo as notas e transformando em dicionario
for i in alunos:   
    dicionario[i] = notas[c]
    c+=1
#criando a variavel com a maior nota    
nota=max(dicionario.values())
#iterando sobre o dicionario 
for i in dicionario:
    #encontrando o Aluno que tirou a maior nota
    if dicionario[i]==nota:
        print ("Nome: {}, nota: {}".format(i,dicionario[i]))
    
```
